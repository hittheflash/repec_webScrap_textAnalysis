```python
import os
import string
import pandas as pd
import time
```

```python
letras = string.ascii_uppercase
total = 0
for i in range(len(letras)):
    archivos = os.listdir(letras[i])
    print("{}: {}".format(letras[i], len(archivos)), end="|")
    total+=len(archivos)
```

```python
total
```

```python
def lista_features(file, lista):
    F = open(file, 'r', encoding="utf8")
    lineas = F.readlines()
    F.close()
    infos = []
    for linea in lineas:
        for tipo in lista:
            if tipo in linea:
                cadena = "{}: ".format(tipo)
                inff = linea.strip().split(cadena)[-1]
                infos.append([tipo, inff])
    return infos
```

```python
def lista_tipos(file, lista):
    F = open(file, 'r', encoding="utf8")
    lineas = F.readlines()
    F.close()
    infos = []
    for linea in lineas:
        for tipo in lista:
            if tipo in linea:
                cadena = "{}: ".format(tipo)
                inff = linea.strip().split(cadena)[-1]
                infos.append([tipo, inff])
    return infos
```

```python
features = ['Name-First', 'Name-Middle', 'Name-Last', 'Name-Suffix', 'Name-Full', 'Name-ASCII', 'Workplace-Name', 'Workplace-Name-English', 'Workplace-Institution', 'Workplace-Share', 'Workplace-Location', 'Workplace-Homepage', 'Email', 'Homepage', 'Postal', 'Phone', 'Short-Id', 'Handle', 'Registered-Date', 'Last-Login-Date', 'Deceased']
tipos = ['author-article', 'author-book', 'author-chapter', 'author-paper', 'author-software', 'editor-book', 'editor-series']
```

```python
t1 = time.time()
print(time.ctime())
DATA = []
for letra in letras:
    print(letra, end="|")
    archivos = os.listdir(letra)
    for archivo in archivos:
        dirr = "{}/{}".format(letra, archivo)
        datos = ["" for i in range(len(features))]

        list_features = lista_features(dirr, features)
        for i in range(len(features)):
            for lista in list_features:
                header, feature = lista
                if features[i]==header:
                    datos[i]=feature

        list_tipos = lista_tipos(dirr, tipos)

        first = datos[:16]
        last = datos[16:]
        for f_tipo in list_tipos:
            general = first + f_tipo  + last
            DATA.append(general)
t2 = time.time()
print()
print(time.ctime())
```

```python
encabezado_general = features[:16] + ["tipo", "reped_code"] + features[16:]
df = pd.DataFrame(DATA)
df.columns = encabezado_general
df
```
