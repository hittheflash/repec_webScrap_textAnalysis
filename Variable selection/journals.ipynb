```python
def nomu(long):
    if '"' in long[0]:
        long[0] = long[0][1:]
    if '"' in long[-1]:
        long[-1] = long[-1][:-2]
    b = long[1:-1]
    c = [i.split(" & ") for i in b]
    d = [";".join(i) for i in c]
    e = "".join(d)
    f = "".join([long[0],e,long[-1]])
    g = f.split(";")
    return g
```

```python
import requests 
from requests import get
from bs4 import BeautifulSoup

def urlsJournals(journal):
    listaJ = []
    www = 'https://ideas.repec.org/a/' + journal
    response = get(www)
    html_soup = BeautifulSoup(response.text, 'html.parser')
    for link in html_soup.find_all("a"):
        urr = link.get("href")
        parte1 = urr.split("y")
        if len(parte1)==2:
            parte2 = parte1[1].split("i")
            anio = int(parte2[0])
            if anio>2009:
                listaJ.append(www + "/" + urr)
    return listaJ
```

```python
import requests 
from requests import get
from bs4 import BeautifulSoup

def generatecsv(dirr, journal):
    name = dirr.split("/")[5]
    response = get(dirr)
    html_content = requests.get(dirr).text.encode("utf-8")
    titulo = BeautifulSoup(html_content, "lxml").title.text
    html_soup = BeautifulSoup(response.text, 'html.parser')
    posts = html_soup.find_all('div', class_= 'tab-pane fade')
    nombres = html_soup.find_all('li', class_= 'authorname')
    nombres = []
    for link in html_soup.find_all("script"):
        cadena = "Inner Text: {}".format(link.text)
        if "schema.org" in cadena:
            part = cadena.split('author": ')[1].split(",")
            idx = part.index('\n\t    "isPartOf": [\n\t      {\n\t        "@id": "#volume"\n\t      }')
            nms = part[:idx]
            if len(nms)==2:
                nam = "".join(nms)
                nombres.append(nam)
            elif len(nms)>2:
                h = nomu(nms)
                for u in h:
                    nombres.append(u)
    post_one = posts[0]
    repec = post_one.input
    repec = str(repec).split("value=")[1].split("/")[0][1:-1]
    parti = repec.split(":")
    year = parti[-5]
    vol = parti[4]
    issue = parti[-3]
    pages = parti[-1].split("-")
    pag0 = pages[0]
    pag1 = pages[1]
    Y = ["HANDLE,TITLE,JOURNAL,YEAR,VOL,ISSUE,PAGE,END PAGE,AUTHORS\n"]
    for autor in nombres:
        linea = "{},{},{},{},{},{},{},{},{}\n".format(repec, titulo, journal, year, vol, issue, pag0, pag1, autor)
        print(linea)
        Y.append(linea)
    file = open(name + "g.csv", "a+")
    file.write(Y[0])
    for cad in Y:
         file.write(cad)
    file.close()
```
